## Indexing and Slicing



### One-dimensional array

Define a 1-D array with values ranging between 0-9.

```python
arr = np.arange(10)
arr
```

Select the first five elements:

```python
arr[:5]
```

Select the elements between the 3rd and 7th indices:

```python
arr[3:7]
```

Select every other element, starting at index 1:

```python
arr[1::2]
```

### Two-dimensional array

Define a 2-D array 4x6 array, with the sequence of values from 1 to 24.

```python
arr = np.arange(1,25).reshape(4,6)
arr
```

Select the first two rows, first three columns:

```python
arr[:2,:3]
```

Select the last three rows, first 2 colums:

```python
arr[1:, :2]
```

Select every other row starting from index 2:

```python
arr[2::2]
```

Select all rows, every other column:

```python
arr[:, ::2]
```

## Boolean Indexing

```python
students = np.array(['A', 'C', 'C', 'B', 'A', 'A', 'C', 'B', 'B'])
students
```

```python
np.random.seed(10)
quizTrials = np.random.rand(9,4)
quizTrials
```

Get the quiz trial data for student 'B'

```python
quizTrials[students == 'B']
```

Get the quiz #3 data for student 'A'

```python
quizTrials[students == 'A', 2]
```

Calculate the quiz score differences between student 'A' and 'C':

```python
diffAC = quizTrials[students == 'A'] - quizTrials[students == 'C']
diffAC
```

Change the negative values to -1, positive values to 1:

```python
diffAC[diffAC > 0] = 1
diffAC[diffAC < 0] = -1
diffAC
```

## Expressing Conditional Logic as Array Operations

```python
students = np.array(['A', 'C', 'C', 'B', 'A', 'A', 'C', 'B', 'B'])
students
```

```python
np.random.seed(10)
quizTrials = np.random.rand(9,4)
quizTrials
```

Mark quiz scores which are above 0.5 as 1 and the rest as 0.

```python
quizTrials = np.where(quizTrials > 0.5, 1, 0)
quizTrials
```

FInd the number times student B got above 0.5 in the last quiz:

```python
quizTrials[students == 'B', 3].sum()
```
