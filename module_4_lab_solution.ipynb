## Generating Dataset 

Our dataset will be page view dataset where column represents the days (DAY 1 to DAY 5), and the row represents the students who visits the pages.

#### TASK 1 [+4]. 
Generate a random array (*named* `data`) with values ranging between 0-1000. The array should have 10 rows and 5 columns. 

**Use the seed of 10** so that it will always generate the same set of random numbers.

```python
np.random.seed(10)
data = np.random.randint(1000, size=(10,5))
data
```

Create a list (or array), named `students`, that contain the following values in the exact order: 
`'C', 'H','A', 'D', 'E', 'J', 'G', 'I', 'B', 'F'`

```python
students = np.array(['C', 'H','A', 'D', 'E', 'J', 'G', 'I', 'B', 'F'])
students
```

Create a new dataframe, which has the `data` as the data source, `students` as the index, and `['Day3', 'Day1', 'Day2', 'Day5', 'Day4']` as the columns. 

```python
visitFrame = pd.DataFrame(data, index=['C', 'H','A', 'D', 'E', 'J', 'G', 'I', 'B', 'F'], columns=['Day3', 'Day1', 'Day2', 'Day5', 'Day4'])
visitFrame
```

## Ordering
As you might have noticed that the order of the row indices and columns are a bit mixed up. You are expected to order both in an ascending order.

#### TASK 2 [+2]. 
Sort the dataframe based on row index.

```python
visitFrame = visitFrame.sort_index()
visitFrame
```

#### TASK 3 [+2]

```python
visitFrame = visitFrame.sort_index(axis=1)
visitFrame
```
