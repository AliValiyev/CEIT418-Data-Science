## Boolean Indexing

Assume that the rows in pageViews array correspond to specific students given in the following array:

```python
students = np.array(['A', 'B','C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])
students
```

#### TASK 5 [+6]. 
Find the visits of students D and H:

```python
pageVisits[((students=='D') | (students=='H'))]
```

#### TASK 6 [+8]. 
Find the visits of students B and F during the first three days:

```python
pageVisits[((students=='B') | (students=='F')), :3]
```

#### TASK 7 [+10]. 
Find all students' visits except student C and H:

```python
exceptC = pageVisits[~((students == 'C') | (students == 'H'))]
exceptC
```

#### TASK 10 [+15]. 
Return the name of the student (which is a letter) who have the lowest average page views. Here you should use [argmin](https://numpy.org/doc/stable/reference/generated/numpy.argmin.html) function.

```python
students[np.argmin(np.average(pageVisits, axis=1))]
```

#### TASK 11 [+15]. 
Find the number of days student A visited more pages than student D.

```python
pvA = pageVisits[students == 'A']
pvD = pageVisits[students == 'D']
diffAD = pvA > pvD
diffAD.sum()
```

#### TASK 12 [+20]. 

Return the list of students whose total visits until the 4th day (that is 1st, 2nd and 3rd days) is less than their visits on the 4th day.

```python
cumSum = np.cumsum(pageVisits, axis=1)
cumSum
```
