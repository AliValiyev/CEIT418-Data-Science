## Task 2

The second task is to find the students whose visits are above the overall average of the whole class. 

Let's assume that the overall average is 50 (*this is just a random number I chose to give you an example*). 

Then, your task is to identify the students who visited module 1, 2, and 3 tutorial pages more than 50 times. 

Overall average should be computed considering all visits (across all modules) by all students.

Your code should produce the following student id numbers:

[1, 3, 5]

````python
allVisits = tutorialModule1 + tutorialModule2 + tutorialModule3
average = sum(allVisits) / len(allVisits)
average
````

````python
[a for a,b,c,d in list(zip(studentIds, tutorialModule1, tutorialModule2, tutorialModule3)) if((b>average) & (c>average) & (d>average))]
````
